## Question 1: Compare and contrast NLTK and spaCy in terms of features, ease of use, and performance.


Here is a clear comparison between **NLTK** and **spaCy**:

---

### 🔹 1. Features

**NLTK (Natural Language Toolkit):**

* Rich collection of linguistic resources (corpora, lexical databases like WordNet).
* Strong support for education and research.
* Provides tokenization, stemming, lemmatization, POS tagging, parsing, and text classification tools.
* More focused on traditional NLP techniques.

**spaCy:**

* Industrial-strength NLP library designed for production use.
* Provides fast tokenization, POS tagging, Named Entity Recognition (NER), dependency parsing, and word vectors.
* Built-in pre-trained deep learning models.
* Easily integrates with ML frameworks.

---

### 🔹 2. Ease of Use

**NLTK:**

* Beginner-friendly and widely used for learning NLP concepts.
* Requires more manual preprocessing and configuration.
* Slower for large-scale production tasks.

**spaCy:**

* Cleaner API and ready-to-use pipelines.
* Minimal setup for advanced tasks like NER and dependency parsing.
* More suitable for real-world applications.

---

### 🔹 3. Performance

**NLTK:**

* Slower because it is mostly Python-based.
* Better suited for small projects, experimentation, and teaching.

**spaCy:**

* Much faster (implemented in Cython).
* Optimized for large datasets and production environments.

---

### 🔹 Summary

| Aspect             | NLTK                 | spaCy                  |
| ------------------ | -------------------- | ---------------------- |
| Primary Focus      | Education & research | Industry & production  |
| Speed              | Slower               | Faster                 |
| Pre-trained Models | Limited              | Strong built-in models |
| Best For           | Learning NLP basics  | Real-world NLP systems |

**In short:**
Use **NLTK** for learning and research; choose **spaCy** for performance-driven, production-level NLP applications.

## Question 2: What is TextBlob and how does it simplify common NLP tasks like sentiment analysis and translation?


**TextBlob** is a high-level Python library built on top of NLTK and Pattern that simplifies common Natural Language Processing (NLP) tasks through an easy-to-use API. It is designed for beginners and rapid prototyping.

### 🔹 How TextBlob Simplifies NLP Tasks

**1. Sentiment Analysis**
TextBlob provides built-in sentiment analysis without requiring manual model training.
It returns:

* **Polarity** (range: -1 to +1, negative to positive sentiment)
* **Subjectivity** (range: 0 to 1, objective to subjective text)

Example:

```python
from textblob import TextBlob

text = TextBlob("NLP is amazing and very powerful!")
print(text.sentiment)
```

Output:

```
Sentiment(polarity=0.75, subjectivity=0.9)
```

---

**2. Translation**
TextBlob can translate text between languages using simple function calls.

Example:

```python
text = TextBlob("Hello, how are you?")
print(text.translate(to='fr'))
```

Output:

```
Bonjour, comment ça va?
```

---

### 🔹 Other Features

* Tokenization
* POS tagging
* Noun phrase extraction
* Spell correction

---

### 🔹 Why It’s Useful

* Very simple syntax
* No complex model setup
* Ideal for small projects and quick experimentation

**In summary:**
TextBlob abstracts complex NLP processes into simple function calls, making tasks like sentiment analysis and translation easy and accessible.

## Question 4: Describe the architecture and functioning of a Recurrent Natural Network (RNN).


A **Recurrent Neural Network (RNN)** is a type of neural network designed to process **sequential data** such as text, speech, or time-series signals. Unlike traditional feedforward networks, RNNs have loops that allow information to persist across time steps.

---

## 🔹 Architecture of an RNN

An RNN consists of:

1. **Input Layer (xₜ)** – Takes the input at time step *t* (e.g., a word in a sentence).
2. **Hidden State (hₜ)** – Stores information from previous time steps. It acts as the network’s memory.
3. **Output Layer (yₜ)** – Produces the prediction at each time step.

The key feature is the **recurrent connection**, where the hidden state from the previous step (hₜ₋₁) is fed back into the network along with the current input.

Mathematically:
$$
h_t = f(W_h h_{t-1} + W_x x_t + b)
$$
$$y_t = g(W_y h_t)
$$

---

## 🔹 How It Works

1. The first input enters the network and produces a hidden state.
2. This hidden state is passed to the next time step along with the new input.
3. The process repeats, allowing the model to capture dependencies across sequences.
4. The final output depends on both current input and previous context.

---

## 🔹 Key Characteristics

* Handles **variable-length sequences**
* Maintains **memory of previous inputs**
* Suitable for tasks like language modeling, sentiment analysis, and speech recognition

---

## 🔹 Limitations

* **Vanishing and exploding gradients** during training
* Difficulty capturing long-term dependencies

To overcome these issues, advanced variants like **LSTM (Long Short-Term Memory)** and **GRU (Gated Recurrent Unit)** were introduced.

---

### ✅ In summary:

An RNN processes sequential data by maintaining a hidden memory state that carries information across time steps, enabling it to model temporal or contextual dependencies in language and other sequence-based tasks.

## Question 5: What is the key difference between LSTM and GRU networks in NLP applications?

The **key difference between LSTM and GRU networks** lies in their internal structure and gating mechanisms used to control information flow in sequential data.

---

### 🔹 LSTM (Long Short-Term Memory)

* Has **three gates**:

  1. **Input gate** – Controls new information added to memory
  2. **Forget gate** – Decides what information to discard
  3. **Output gate** – Determines what part of memory to output
* Maintains **two states**:

  * Cell state (long-term memory)
  * Hidden state (short-term memory)
* More complex and powerful for capturing long-range dependencies.

---

### 🔹 GRU (Gated Recurrent Unit)

* Has **two gates**:

  1. **Update gate**
  2. **Reset gate**
* Combines cell state and hidden state into a **single state**
* Simpler architecture with fewer parameters
* Faster to train and computationally efficient.

---

### 🔹 Key Difference Summary

| Aspect          | LSTM                         | GRU                 |
| --------------- | ---------------------------- | ------------------- |
| Number of Gates | 3                            | 2                   |
| Memory States   | Separate cell + hidden state | Single hidden state |
| Complexity      | More complex                 | Simpler             |
| Training Speed  | Slower                       | Faster              |

---

### ✅ Conclusion

LSTMs are more expressive and better for complex long-term dependencies, while GRUs are simpler, faster, and often perform similarly with fewer parameters in many NLP tasks.

## Question 6: Write a Python program using TextBlob to perform sentiment analysis on the following paragraph of text:
## “I had a great experience using the new mobile banking app. The interface is intuitive, and customer support was quick to resolve my issue. However, the app did crash once during a transaction, which was frustrating"

Your program should print out the polarity and subjectivity scores.

(Include your Python code and output in the code box below.)



````python

from textblob import TextBlob

text = """I had a great experience using the new mobile banking app.
The interface is intuitive, and customer support was quick to resolve my issue.
However, the app did crash once during a transaction, which was frustrating."""

blob = TextBlob(text)

# Sentiment Analysis
polarity = blob.sentiment.polarity
subjectivity = blob.sentiment.subjectivity

print("Polarity Score:", polarity)
print("Subjectivity Score:", subjectivity)
````

## Question 7: Given the sample paragraph below, perform string tokenization and frequency distribution using Python and NLTK:
## “Natural Language Processing (NLP) is a fascinating field that combines linguistics, computer science, and artificial intelligence. It enables machines to understand, interpret, and generate human language. Applications of NLP include chatbots, sentiment analysis, and machine translation. As technology advances, the role of NLP in modern solutions is becoming increasingly critical.”

(Include your Python code and output in the code box below.)

````python

import nltk
from nltk.tokenize import word_tokenize
from nltk.probability import FreqDist

# Download tokenizer (run once)
nltk.download('punkt_tab')

text = """Natural Language Processing (NLP) is a fascinating field that combines linguistics,
computer science, and artificial intelligence. It enables machines to understand,
interpret, and generate human language. Applications of NLP include chatbots,
sentiment analysis, and machine translation. As technology advances, the role of NLP
in modern solutions is becoming increasingly critical."""

# Tokenization
tokens = word_tokenize(text)

# Convert to lowercase and remove punctuation
words = [word.lower() for word in tokens if word.isalpha()]

# Frequency Distribution
freq_dist = FreqDist(words)

print("Tokens:")
print(words)

print("\nFrequency Distribution:")
for word, freq in freq_dist.items():
    print(f"{word}: {freq}")
````

## Question 8: Implement a basic LSTM model in Keras for a text classification task using the following dummy dataset. Your model should classify sentences as either positive (1) or negative (0).

## Dataset
texts = [

“I love this project”, #Positive
“This is an amazing experience”, #Positive
“I hate waiting in line”, #Negative
“This is the worst service”, #Negative
“Absolutely fantastic!” #Positive

]

labels = [1, 1, 0, 0, 1]

### Preprocess the text, tokenize it, pad sequences, and build an LSTM model to train on this data. You may use Keras with TensorFlow backend.

(Include your Python code and output in the code box below.)


````python

import numpy as np
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense

# -----------------------
# 1. Dummy Dataset
# -----------------------
texts = [
    "I love this project",
    "This is an amazing experience",
    "I hate waiting in line",
    "This is the worst service",
    "Absolutely fantastic"
]

labels = [1, 1, 0, 0, 1]

# -----------------------
# 2. Tokenization
# -----------------------
tokenizer = Tokenizer(num_words=1000)
tokenizer.fit_on_texts(texts)

sequences = tokenizer.texts_to_sequences(texts)

# -----------------------
# 3. Padding
# -----------------------
max_length = 6
X = pad_sequences(sequences, maxlen=max_length, padding='post')

y = np.array(labels)

# -----------------------
# 4. Build LSTM Model
# -----------------------
model = Sequential([
    Embedding(input_dim=1000, output_dim=16, input_length=max_length),
    LSTM(32),
    Dense(1, activation='sigmoid')
])

model.compile(loss='binary_crossentropy',
              optimizer='adam',
              metrics=['accuracy'])

# -----------------------
# 5. Train Model
# -----------------------
model.fit(X, y, epochs=20, verbose=1)

# -----------------------
# 6. Evaluate
# -----------------------
loss, accuracy = model.evaluate(X, y, verbose=0)

print("\nFinal Loss:", loss)
print("Final Accuracy:", accuracy)
````
