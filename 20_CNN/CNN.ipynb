### Question 1: What is the role of filters and feature maps in Convolutional Neural Network (CNN)?

In a **Convolutional Neural Network (CNN)**, **filters (kernels)** and **feature maps** are core components that enable the model to automatically extract meaningful patterns from input data.

* **Filters (Kernels):** Small learnable matrices that slide over the input image to detect local patterns such as edges, corners, and textures. Each filter focuses on a specific type of feature and its values are learned during training.

* **Feature Maps:** The outputs produced after applying filters across the input. Each feature map highlights where a particular feature appears in the image, preserving spatial relationships.

**In essence**, filters act as **pattern detectors**, and feature maps represent the **presence and location of those patterns**, enabling CNNs to build hierarchical visual understanding.

### Question 2: Explain the concepts of padding and stride in CNNs(Convolutional Neural Network). How do they affect the output dimensions of feature maps?

In **Convolutional Neural Networks (CNNs)**, **padding** and **stride** control how the convolution operation moves over the input and directly influence the **size of the output feature maps**.

**Padding:**
Padding adds extra pixels (usually zeros) around the input image.

* **Purpose:** Prevents loss of information at image borders and helps control output size.
* **Effect:**

  * *Valid padding* (no padding) → output size decreases.
  * *Same padding* → output size remains the same as input.

**Stride:**
Stride defines the step size with which the filter moves across the input.

* **Purpose:** Controls spatial downsampling.
* **Effect:**

  * Larger stride → smaller output feature map.
  * Smaller stride → larger, more detailed feature map.

**Output size formula:**
$$\text{Output} = \frac{(N - F + 2P)}{S} + 1$$
where (N) = input size, (F) = filter size, (P) = padding, and (S) = stride.

**In summary**, padding preserves spatial information, while stride controls resolution and computational efficiency.


##
### Question 3: Define receptive field in the context of CNNs. Why is it important for deep architectures?

In **Convolutional Neural Networks (CNNs)**, the **receptive field** refers to the **region of the input image that influences a particular neuron or feature map activation**.

**Importance in deep architectures:**

* As network depth increases, the **effective receptive field grows**, allowing neurons to capture **larger and more complex patterns**.
* Early layers focus on **local features** (edges, textures), while deeper layers learn **global structures** (objects, shapes).
* A larger receptive field enables better **context understanding**, which is crucial for tasks like image classification and object detection.

**In essence**, receptive fields allow CNNs to build hierarchical feature representations from local to global information.


##
### Question 4: Discuss how filter size and stride influence the number of parameters in a CNN.

In a **Convolutional Neural Network (CNN)**, **filter size** and **stride** affect the model’s **parameter count and computational cost**, though in different ways.

**Filter size:**

* Larger filters (e.g., 5×5 vs 3×3) have **more trainable parameters** per filter.
* Number of parameters per filter = (F \times F \times C) (where (C) is input channels).
* Larger filters capture wider context but increase **model complexity and risk of overfitting**.

**Stride:**

* Stride does **not change the number of parameters**, since weights remain the same.
* Larger stride reduces the **number of feature map activations**, lowering computation and memory usage.

**Summary:**
Filter size directly increases parameters and model capacity, while stride controls spatial resolution and efficiency without affecting parameter count.

### Question 5: Compare and contrast different CNN-based architectures like LeNet, AlexNet, and VGG in terms of depth, filter sizes, and performance.

Below is a **clear comparison** of **LeNet, AlexNet, and VGG** based on **architecture depth, filter design, and performance** 👇

### **Comparison of CNN Architectures**

| **Aspect**              | **LeNet**                | **AlexNet**                      | **VGG**                               |
| ----------------------- | ------------------------ | -------------------------------- | ------------------------------------- |
| **Introduced**          | 1998                     | 2012                             | 2014                                  |
| **Depth**               | Shallow (≈5 layers)      | Medium (8 layers)                | Deep (16–19 layers)                   |
| **Filter Sizes**        | Large (5×5)              | Mixed (11×11, 5×5, 3×3)          | Uniform small (3×3)                   |
| **Activation Function** | Tanh / Sigmoid           | ReLU                             | ReLU                                  |
| **Pooling**             | Average Pooling          | Max Pooling                      | Max Pooling                           |
| **Parameters**          | Very few                 | ~60 million                      | Very high (~138 million)              |
| **Performance**         | Simple digit recognition | ImageNet breakthrough            | High accuracy, heavy computation      |
| **Use Case**            | MNIST-like tasks         | Large-scale image classification | Feature extraction, transfer learning |

### ✅ Key Takeaways

* **LeNet** is lightweight and suited for simple tasks.
* **AlexNet** introduced deep learning success with ReLU and GPUs.
* **VGG** emphasized depth using small filters, improving accuracy at the cost of computation.

### Question 6: Using keras, build and train a simple CNN model on the MNIST dataset from scratch. Include code for module creation, compilation, training, and evaluation.

```python
# --- Suppress TensorFlow Warnings ---
import os, warnings
os.environ['TF_CPP_MIN_LOG_LEVEL'] = '3'
os.environ['TF_ENABLE_ONEDNN_OPTS'] = '0'
warnings.filterwarnings("ignore")

# --- Import Libraries ---
import tensorflow as tf
from tensorflow.keras.datasets import mnist
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense
from tensorflow.keras.utils import to_categorical

# --- Load MNIST Dataset ---
(X_train, y_train), (X_test, y_test) = mnist.load_data()

# --- Preprocess Data ---
X_train = X_train.reshape(-1, 28, 28, 1) / 255.0
X_test = X_test.reshape(-1, 28, 28, 1) / 255.0

y_train = to_categorical(y_train, 10)
y_test = to_categorical(y_test, 10)

# --- Build CNN Model ---
model = Sequential([
    Conv2D(32, (3, 3), activation='relu', input_shape=(28, 28, 1)),
    MaxPooling2D(pool_size=(2, 2)),

    Conv2D(64, (3, 3), activation='relu'),
    MaxPooling2D(pool_size=(2, 2)),

    Flatten(),
    Dense(128, activation='relu'),
    Dense(10, activation='softmax')
])

# --- Compile Model ---
model.compile(
    optimizer='adam',
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

# --- Train Model ---
model.fit(
    X_train, y_train,
    epochs=5,
    batch_size=128,
    validation_split=0.1,
    verbose=1
)

# --- Evaluate Model ---
test_loss, test_accuracy = model.evaluate(X_test, y_test, verbose=0)
print(f"Test Accuracy: {test_accuracy:.4f}")
```

### Question 7: Load and preprocess the CIFAR-10 dataset using Keras, and create a CNN model to classify RGB images. Show your preprocessing and architecture.

```python
# --- Suppress TensorFlow Warnings ---
import os, warnings
os.environ['TF_CPP_MIN_LOG_LEVEL'] = '3'
os.environ['TF_ENABLE_ONEDNN_OPTS'] = '0'
warnings.filterwarnings("ignore")

# --- Import Libraries ---
import tensorflow as tf
from tensorflow.keras.datasets import cifar10
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout
from tensorflow.keras.utils import to_categorical

# --- Load CIFAR-10 Dataset ---
(X_train, y_train), (X_test, y_test) = cifar10.load_data()

# --- Preprocessing ---
# Normalize pixel values (0–255 → 0–1)
X_train = X_train / 255.0
X_test = X_test / 255.0

# One-hot encode labels
y_train = to_categorical(y_train, 10)
y_test = to_categorical(y_test, 10)

print("Training data shape:", X_train.shape)
print("Test data shape:", X_test.shape)

# --- Build CNN Model ---
model = Sequential([
    Conv2D(32, (3, 3), activation='relu', padding='same', input_shape=(32, 32, 3)),
    MaxPooling2D(pool_size=(2, 2)),

    Conv2D(64, (3, 3), activation='relu', padding='same'),
    MaxPooling2D(pool_size=(2, 2)),

    Conv2D(128, (3, 3), activation='relu', padding='same'),
    MaxPooling2D(pool_size=(2, 2)),

    Flatten(),
    Dense(256, activation='relu'),
    Dropout(0.5),
    Dense(10, activation='softmax')
])

# --- Compile Model ---
model.compile(
    optimizer='adam',
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

# --- Model Summary ---
model.summary()

# --- Train Model ---
model.fit(
    X_train, y_train,
    epochs=10,
    batch_size=128,
    validation_split=0.1,
    verbose=1
)

# --- Evaluate Model ---
test_loss, test_accuracy = model.evaluate(X_test, y_test, verbose=0)
print(f"Test Accuracy: {test_accuracy:.4f}")
```

### Question 8: Using PyTorch, write a script to define and train a CNN on the MNIST dataset. Include model definition, data loaders, training loop, and accuracy evaluation.

```python
# --- Import Libraries ---
import torch
import torch.nn as nn
import torch.optim as optim
from torchvision import datasets, transforms
from torch.utils.data import DataLoader

# --- Device Configuration ---
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("Using device:", device)

# --- Data Preprocessing ---
transform = transforms.Compose([
    transforms.ToTensor(),
    transforms.Normalize((0.5,), (0.5,))
])

# --- Load MNIST Dataset ---
train_dataset = datasets.MNIST(
    root="./data",
    train=True,
    transform=transform,
    download=True
)

test_dataset = datasets.MNIST(
    root="./data",
    train=False,
    transform=transform,
    download=True
)

train_loader = DataLoader(train_dataset, batch_size=64, shuffle=True)
test_loader = DataLoader(test_dataset, batch_size=64, shuffle=False)

# --- Define CNN Model ---
class CNN(nn.Module):
    def __init__(self):
        super(CNN, self).__init__()

        self.conv1 = nn.Conv2d(1, 32, kernel_size=3)
        self.conv2 = nn.Conv2d(32, 64, kernel_size=3)
        self.pool = nn.MaxPool2d(2, 2)
        self.fc1 = nn.Linear(64 * 5 * 5, 128)
        self.fc2 = nn.Linear(128, 10)
        self.relu = nn.ReLU()

    def forward(self, x):
        x = self.pool(self.relu(self.conv1(x)))
        x = self.pool(self.relu(self.conv2(x)))
        x = x.view(x.size(0), -1)
        x = self.relu(self.fc1(x))
        x = self.fc2(x)
        return x

model = CNN().to(device)

# --- Loss Function and Optimizer ---
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)

# --- Training Loop ---
epochs = 5

for epoch in range(epochs):
    model.train()
    running_loss = 0.0

    for images, labels in train_loader:
        images, labels = images.to(device), labels.to(device)

        optimizer.zero_grad()
        outputs = model(images)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()

        running_loss += loss.item()

    print(f"Epoch [{epoch+1}/{epochs}], Loss: {running_loss/len(train_loader):.4f}")

# --- Evaluation ---
model.eval()
correct = 0
total = 0

with torch.no_grad():
    for images, labels in test_loader:
        images, labels = images.to(device), labels.to(device)
        outputs = model(images)
        _, predicted = torch.max(outputs, 1)
        total += labels.size(0)
        correct += (predicted == labels).sum().item()

accuracy = 100 * correct / total
print(f"Test Accuracy: {accuracy:.2f}%")
```

### Question 9: Given a custom image dataset stored in a local directory, write code using Keras ImageDataGenerator to preprocess and train a CNN model.

```python
import tensorflow as tf
import tensorflow_datasets as tfds

# Load dataset directly with TensorFlow Datasets
(ds_train, ds_val), ds_info = tfds.load(
    'cats_vs_dogs',
    split=['train[:80%]', 'train[80%:90%]'],
    as_supervised=True,
    with_info=True
)

def preprocess(image, label):
    image = tf.image.resize(image, (150,150)) / 255.0
    return image, label

train_dataset = ds_train.map(preprocess).batch(32)
val_dataset = ds_val.map(preprocess).batch(32)

# Build model (same as before)
model = tf.keras.Sequential([
    tf.keras.layers.Conv2D(32, (3,3), activation="relu", input_shape=(150,150,3)),
    tf.keras.layers.MaxPool2D(2,2),
    tf.keras.layers.Conv2D(64, (3,3), activation="relu"),
    tf.keras.layers.MaxPool2D(2,2),
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(128, activation="relu"),
    tf.keras.layers.Dropout(0.5),
    tf.keras.layers.Dense(1, activation="sigmoid")
])

model.compile(optimizer="adam",
              loss="binary_crossentropy",
              metrics=["accuracy"])

history = model.fit(train_dataset, epochs=5, validation_data=val_dataset)

val_loss, val_acc = model.evaluate(val_dataset)
print(f"Validation Accuracy: {val_acc:.4f}")
```

```python
# =========================================================
# STEP 0: SUPPRESS WARNINGS
# =========================================================
import os, warnings
os.environ['TF_CPP_MIN_LOG_LEVEL'] = '3'
os.environ['TF_ENABLE_ONEDNN_OPTS'] = '0'
warnings.filterwarnings("ignore")

# =========================================================
# STEP 1: IMPORT LIBRARIES
# =========================================================
import tensorflow as tf
import tensorflow_datasets as tfds
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout

# =========================================================
# STEP 2: LOAD DATASET
# =========================================================
(ds_train, ds_val), ds_info = tfds.load(
    'cats_vs_dogs',
    split=['train[:10%]', 'train[10%:15%]'],  # small fractions for demo
    as_supervised=True,
    with_info=True
)

def preprocess(image, label):
    image = tf.image.resize(image, (150,150)) / 255.0
    return image, label

train_data = ds_train.map(preprocess).batch(32).prefetch(tf.data.AUTOTUNE)
val_data = ds_val.map(preprocess).batch(32).prefetch(tf.data.AUTOTUNE)

# =========================================================
# STEP 3: BUILD CNN MODEL
# =========================================================
model = Sequential([
    Conv2D(32, (3,3), activation="relu", input_shape=(150,150,3)),
    MaxPooling2D(2,2),

    Conv2D(64, (3,3), activation="relu"),
    MaxPooling2D(2,2),

    Conv2D(128, (3,3), activation="relu"),
    MaxPooling2D(2,2),

    Flatten(),
    Dense(128, activation="relu"),
    Dropout(0.5),
    Dense(1, activation="sigmoid")
])

model.compile(
    optimizer="adam",
    loss="binary_crossentropy",
    metrics=["accuracy"]
)

# =========================================================
# STEP 4: TRAIN MODEL
# =========================================================
history = model.fit(train_data, epochs=3, validation_data=val_data)

# Save trained model
model.save("cnn_model.h5")

# =========================================================
# STEP 5: STREAMLIT DEPLOYMENT
# =========================================================
# Run this section using: streamlit run filename.py
import streamlit as st
import numpy as np
from PIL import Image

st.title("Cats vs Dogs Classifier (Demo)")

@st.cache_resource
def load_model():
    return tf.keras.models.load_model("cnn_model.h5")

model = load_model()

uploaded_file = st.file_uploader("Upload an Image", type=["jpg","png"])
if uploaded_file:
    img = Image.open(uploaded_file).resize((150,150))
    img_array = np.array(img)/255.0
    img_array = img_array.reshape(1,150,150,3)

    prediction = model.predict(img_array)[0][0]
    label = "Dog" if prediction > 0.5 else "Cat"

    st.image(img, caption="Uploaded Image")
    st.subheader(f"Prediction: {label}")
    st.write(f"Confidence Score: {prediction:.2f}")
```
