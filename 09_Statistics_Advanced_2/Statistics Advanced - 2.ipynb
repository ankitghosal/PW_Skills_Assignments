### Question 1: What is hypothesis testing in statistics?


**Hypothesis testing** is a statistical method used to make decisions or draw conclusions about a population based on sample data. It involves formulating two competing statements:

* **Null Hypothesis (Hâ‚€):** Assumes no effect or no difference exists (status quo).
* **Alternative Hypothesis (Hâ‚):** Assumes there is an effect or difference.

The test uses sample data and probability theory to determine whether to **reject Hâ‚€** in favor of Hâ‚, or **fail to reject Hâ‚€**.

ðŸ‘‰ It is widely used in decision-making, research validation, and comparing groups (e.g., testing if a new product increases sales).


####
### Question 2: What is the null hypothesis, and how does it differ from the alternative hypothesis?

* **Null Hypothesis (Hâ‚€):** A baseline assumption that there is **no effect, no difference, or no relationship** in the population. Example: â€œThe new drug has no effect on blood pressure compared to the old drug.â€

* **Alternative Hypothesis (Hâ‚ or Ha):** A competing claim that there **is an effect, a difference, or a relationship**. Example: â€œThe new drug lowers blood pressure more than the old drug.â€

ðŸ‘‰ **Difference:** Hâ‚€ represents the status quo or default assumption, while Hâ‚ represents the research claim being tested. Hypothesis testing evaluates whether the evidence is strong enough to reject Hâ‚€ in favor of Hâ‚.

### Question 3: Explain the significance level in hypothesis testing and its role in deciding the outcome of a test.


The **significance level (Î±)** is the threshold probability used in hypothesis testing to decide whether to reject the null hypothesis (Hâ‚€). It represents the **maximum acceptable risk of making a Type I error** (rejecting Hâ‚€ when it is actually true).

* Common values: **0.05 (5%)**, **0.01 (1%)**.
* If the **p-value â‰¤ Î±**, we reject Hâ‚€ â†’ evidence supports the alternative hypothesis.
* If the **p-value > Î±**, we fail to reject Hâ‚€ â†’ insufficient evidence against the null.

ðŸ‘‰ In short, Î± sets the **strictness of the test** and directly influences decision-making in hypothesis testing.


####
### Question 4: What are Type I and Type II errors? Give examples of each.

* **Type I Error (False Positive):** Occurs when we **reject the null hypothesis (Hâ‚€) even though it is true**.

  * Example: Concluding that a new medicine works better than the old one when, in reality, it does not.

* **Type II Error (False Negative):** Occurs when we **fail to reject the null hypothesis (Hâ‚€) even though it is false**.

  * Example: Concluding that a new medicine has no effect when, in reality, it actually works better.

ðŸ‘‰ In short, **Type I = false alarm**, **Type II = missed detection**.

### Question 5: What is the difference between a Z-test and a T-test? Explain when to use each.


The **Z-test** and **T-test** are both hypothesis tests used to compare sample data with population parameters, but they differ mainly in assumptions about population variance and sample size:

* **Z-test:** Used when the **population standard deviation (Ïƒ) is known** and the **sample size is large (n > 30)**. It relies on the standard normal distribution (Z-distribution).

  * Example: Testing whether the average height of 10,000 students matches a known population mean when Ïƒ is known.

* **T-test:** Used when the **population standard deviation is unknown** and especially for **small sample sizes (n â‰¤ 30)**. It relies on the **t-distribution**, which has heavier tails to account for added uncertainty.

  * Example: Comparing the mean exam score of a class of 20 students to a population mean without knowing Ïƒ.

ðŸ‘‰ **Rule of thumb:** If Ïƒ is known & n is large â†’ **Z-test**; if Ïƒ is unknown or n is small â†’ **T-test**.

### Question 6: Write a Python program to generate a binomial distribution with n=10 and p=0.5, then plot its histogram.
(Include your Python code and output in the code box below.)
Hint: Generate random number using random function.

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
n = 10   # number of trials
p = 0.5  # probability of success
size = 1000  # number of random samples

# Generate binomial distribution
binomial_data = np.random.binomial(n, p, size)

# Compute basic statistics
mean = np.mean(binomial_data)
std_dev = np.std(binomial_data)

print("Mean:", mean)
print("Standard Deviation:", std_dev)

# Plot histogram
plt.hist(binomial_data, bins=range(n+2), edgecolor='black', alpha=0.7)
plt.title("Binomial Distribution (n=10, p=0.5)")
plt.xlabel("Number of Successes")
plt.ylabel("Frequency")
plt.show()
```

### Question 7: Implement hypothesis testing using Z-statistics for a sample dataset in Python. Show the Python code and interpret the results.
sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
50.3, 50.4, 50.0, 49.7, 50.5, 49.9]
(Include your Python code and output in the code box below.)


```python
import numpy as np
from scipy.stats import norm

# Sample data
sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
50.3, 50.4, 50.0, 49.7, 50.5, 49.9]

# Convert to numpy array
data = np.array(sample_data)

# Hypothesized population mean
mu_0 = 50  

# Calculate sample statistics
sample_mean = np.mean(data)
sample_std = np.std(data, ddof=1)   # sample standard deviation
n = len(data)

# Z-statistic
z_stat = (sample_mean - mu_0) / (sample_std / np.sqrt(n))

# p-value (two-tailed test)
p_value = 2 * (1 - norm.cdf(abs(z_stat)))

print("Sample Mean:", sample_mean)
print("Sample Standard Deviation:", sample_std)
print("Z-statistic:", z_stat)
print("p-value:", p_value)

# Interpretation
alpha = 0.05
if p_value < alpha:
    print("Reject the Null Hypothesis: The sample mean is significantly different from 50.")
else:
    print("Fail to Reject the Null Hypothesis: No significant difference from 50.")
```

### Question 8: Write a Python script to simulate data from a normal distribution and calculate the 95% confidence interval for its mean. Plot the data using Matplotlib.
(Include your Python code and output in the code box below.)


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Step 1: Simulate data from a normal distribution
np.random.seed(42)  # for reproducibility
data = np.random.normal(loc=50, scale=5, size=1000)  # mean=50, std=5, n=1000

# Step 2: Calculate sample mean and standard error
sample_mean = np.mean(data)
sample_std = np.std(data, ddof=1)
n = len(data)
standard_error = sample_std / np.sqrt(n)

# Step 3: 95% confidence interval (using Z=1.96 for 95% CI)
z_value = 1.96
ci_lower = sample_mean - z_value * standard_error
ci_upper = sample_mean + z_value * standard_error

print("Sample Mean:", sample_mean)
print("95% Confidence Interval:", (ci_lower, ci_upper))

# Step 4: Plot histogram of the data
plt.figure(figsize=(7,5))
plt.hist(data, bins=30, color='skyblue', edgecolor='black', density=True, alpha=0.7)

# Plot normal distribution curve
x = np.linspace(min(data), max(data), 100)
plt.plot(x, norm.pdf(x, sample_mean, sample_std), 'r-', lw=2, label='Normal Fit')

# Mark CI boundaries
plt.axvline(ci_lower, color='green', linestyle='--', label=f'95% CI Lower ({ci_lower:.2f})')
plt.axvline(ci_upper, color='green', linestyle='--', label=f'95% CI Upper ({ci_upper:.2f})')
plt.axvline(sample_mean, color='blue', linestyle='-', label=f'Mean ({sample_mean:.2f})')

plt.title("Normal Distribution with 95% Confidence Interval")
plt.xlabel("Values")
plt.ylabel("Density")
plt.legend()
plt.show()
```

### Question 9: Write a Python function to calculate the Z-scores from a dataset and visualize the standardized data using a histogram. Explain what the Z-scores represent
in terms of standard deviations from the mean.
(Include your Python code and output in the code box below.)

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import zscore

# Function to calculate Z-scores and plot histogram
def calculate_zscores(data):
    # Step 1: Compute Z-scores
    z_scores = zscore(data)  # (x - mean) / std
    
    # Step 2: Print first few Z-scores
    print("First 10 Z-scores:", z_scores[:10])
    
    # Step 3: Plot histogram of standardized data
    plt.figure(figsize=(7,5))
    plt.hist(z_scores, bins=30, color='purple', edgecolor='black', alpha=0.7, density=True)
    plt.axvline(0, color='blue', linestyle='-', label='Mean (0)')
    plt.axvline(1, color='green', linestyle='--', label='+1 Std Dev')
    plt.axvline(-1, color='green', linestyle='--', label='-1 Std Dev')
    plt.title("Histogram of Standardized Data (Z-scores)")
    plt.xlabel("Z-score")
    plt.ylabel("Density")
    plt.legend()
    plt.show()
    
    return z_scores

# Example dataset
data = np.random.normal(50, 5, 1000)  # mean=50, std=5
z_scores = calculate_zscores(data)
```
